# Gradient Flow

This is a short notebook to demonstrate certain properties of gradient flow on two simple quadratic functions. In modern deep learning, we deal with highly non-convex functions. Although the scenarios in this notebook doesn't necessarily reflect these conditions, they provide a good basis in understanding the gradient dynamics, certain problems that arise with non-convex functions and how to ease them.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D plotting module

%matplotlib inline
```

We define two quadratic functions: convex and non-convex, as well as their corresponding gradients. The general function is of the form $x^T A x + b^T x + c$.

- **Convex function**: $f(x) = \frac 1 2 \lVert x\rVert^2$, $\nabla f(x) = x$.
- **Non-convex function**: $f(x) = x^T \left[\begin{array}{cc} -1 & 0 \\ 0 & 2\end{array}\right]x + 3$. Note that $A$ has both negative and positive eigen-values, hence we can conclude that $f$ is not convex. $A=\mathbb I$ in the convex case. Please note that the sample convex function here is not bounded, we are using it for demonstrative purposes

```python
# Define the quadratic function: 0.5 * x^T x in this case
# 0.5 * x^T I x
def quadratic_function(x):
    return 0.5 * np.sum(np.square(x), axis=0)

# Define the gradient of the quadratic function
def gradient_quadratic_function(x):
    return x

# Gradient flow function to update x
def q_gradient_flow_step(x, learning_rate):
    gradient = gradient_quadratic_function(x)
    return x - learning_rate * gradient

# Define the nonconvex quadratic function
def nonconvex_example(x):
    return - x[0]**2 + 2*x[1]**2 + 3

def gradient_nonconvex_function(x):
    return np.stack((-2*x[0], 4* x[1]))

def nc_gradient_flow_step(x, learning_rate):
    gradient = gradient_nonconvex_function(x)
    return x - learning_rate * gradient

def convex_functions():
    """ returns the function, its gradient and the gradient flow function"""
    return quadratic_function, gradient_quadratic_function, q_gradient_flow_step


def nonconvex_functions():
    """ returns the function, its gradient and the gradient flow function"""
    return nonconvex_example, gradient_nonconvex_function, nc_gradient_flow_step
```

```python

def plot_steepest_descent_direction(func, grad_func, model_type: str = "Convex", bound: int = 4):
    bounds = np.linspace(-bound, bound, 40)
    fig = plt.figure(figsize=(7, 3.5))
    X1, X2 = np.meshgrid(bounds, bounds)
    mesh = np.stack((X1, X2))
    steepest_descent = -grad_func(mesh)
    steepest_descent = steepest_descent / np.linalg.norm(steepest_descent, axis=1)[:, np.newaxis]

    plt.contourf(X1, X2,
                func(mesh),
                levels=20,
                cmap=plt.cm.jet, alpha=0.6)
    plt.quiver(X1, X2, steepest_descent[0], steepest_descent[1], scale=20, color='blue',
              label='Gradient Direction')
    plt.colorbar()
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(f'Steepest Descent in a {model_type} Quadratic Model')
    plt.legend()

plot_steepest_descent_direction(quadratic_function, gradient_quadratic_function, "Convex")
plot_steepest_descent_direction(nonconvex_example, gradient_nonconvex_function, "NonConvex")
```

```python

def plot_steepest_descent_direction_3d(func, grad_func, model_name: str = "Convex"):
    bounds = np.linspace(-4, 4, 40)

    # Create a grid of points for visualization
    X1, X2 = np.meshgrid(bounds, bounds)
    mesh = np.stack((X1, X2))

    # Calculate function values
    Z = func(mesh)

    # Calculate gradient at each point
    grad = grad_func(mesh)
    grad_norm = np.linalg.norm(grad, axis=0)

    # Create a 3D plot
    fig = plt.figure(figsize=(7, 3.5))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the 3D surface of the function
    ax.plot_surface(X1, X2, Z, cmap=plt.cm.jet, rstride=1, cstride=1, alpha=0.6, linewidth=0, antialiased=False)

    # Plot the gradient flow direction
    ax.quiver(X1, X2, Z, -grad[0], -grad[1], np.zeros_like(Z), color='blue', label='Gradient Direction', normalize=True, length=0.1)

    # Add a color bar which maps values to colors
    cbar = fig.colorbar(ax.plot_surface(X1, X2, Z, cmap=plt.cm.jet, rstride=1, cstride=1, alpha=0.6, linewidth=0, antialiased=False), shrink=0.5, aspect=5)
    cbar.set_label('Function Value')

    # Set labels and title
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('Function Value')
    ax.set_title(f'Stepest Descent in a {model_name} Quadratic Model (3D)')

    # Add a legend
    ax.legend()

    # Show the plot
    plt.show()

# Plot the 3D visualization
plot_steepest_descent_direction_3d(quadratic_function, gradient_quadratic_function, "Convex")
plot_steepest_descent_direction_3d(nonconvex_example, gradient_nonconvex_function, "NonConvex")
```

## Gradient Flow in the Convex Case

```python
func, grad, grad_flow_step = convex_functions()
title = "Convex"

# Parameters
learning_rate = 0.01
num_steps = 900
initial_x = np.array([0, 4.0])  # Initial starting point

# Lists to store the trajectory of x and function values
x_trajectory = [initial_x]
function_values = [func(initial_x)]

# Perform gradient flow steps
for _ in range(num_steps):
    initial_x = grad_flow_step(initial_x, learning_rate)
    x_trajectory.append(initial_x)
    function_values.append(func(initial_x))

# Convert lists to NumPy arrays for easier plotting
x_trajectory = np.array(x_trajectory)
function_values = np.array(function_values)

bounds = np.linspace(-4, 4, 100)
# Plot the trajectory
mesh = np.meshgrid(bounds, bounds)
Z = func(mesh)
plot_steepest_descent_direction(func, grad, title)
plt.contourf(*mesh,
             Z,
             levels=20,
             cmap=plt.cm.jet, alpha=0.6)
plt.scatter(x_trajectory[:, 0], x_trajectory[:, 1], c='red', marker='x', label='Gradient Flow')
plt.plot(x_trajectory[:, 0], x_trajectory[:, 1], c='red', linestyle='dashed', linewidth=0.8)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
surf = ax.plot_surface(*mesh, Z, cmap=plt.cm.jet, rstride=1, cstride=1, alpha=0.8, antialiased=True)
ax.plot(x_trajectory[:, 0], x_trajectory[:, 1], function_values, marker='o', linestyle='-', color='red', label='Gradient Flow Trajectory')

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5)

# Set labels and title
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('Function Value')
ax.set_title('3D Plot of Non-Convex Function')

# plt.colorbar()
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gradient Flow in a Convex Quadratic Model')
plt.legend()
plt.show()
```

Let's simulate random initializations, for the convex case

```python
N_SIMULATIONS = 10
BOUND = 4

func, grad, gradient_flow_step = convex_functions()
learning_rate = 0.1
num_steps = 50

plot_steepest_descent_direction(func, grad)

# Plot the trajectory
for sim in range(N_SIMULATIONS):
    initial_x = np.random.normal(0, 1, 2) # Initial starting point

    # Lists to store the trajectory of x and function values
    x_trajectory = [initial_x]
    function_values = [func(initial_x)]

    # Perform gradient flow steps
    for _ in range(num_steps):
        initial_x = gradient_flow_step(initial_x, learning_rate)
        x_trajectory.append(initial_x)
        function_values.append(func(initial_x))

    # Convert lists to NumPy arrays for easier plotting
    x_trajectory = np.array(x_trajectory)
    function_values = np.array(function_values)

    color = np.random.rand(3)   # pick a random color
    plt.scatter(x_trajectory[:, 0], x_trajectory[:, 1], c=color, marker='x', label=f'Gradient Flow {sim}')
    plt.plot(x_trajectory[:, 0], x_trajectory[:, 1], c=color, linestyle='dashed', linewidth=0.8)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gradient Flow in a Convex Quadratic Model')
plt.show()
```

- Q: Describe the direction of the gradient flow. How does the magnitude change along the trajectory?
- A: It follows the negative gradient, i.e. the direction of stepest descent. As the algorithm gets close to the minimum (flat region) the gradient magnitude decreases and we take smaller steps.

- Q: Play with the learning rate, number of steps. When do we have convergence?  
- A: Since this is a convex function, there is a global minimum, which gradient descent is able to find in finite number of steps. However note that if we use a small learning rate and little number of steps we might end up in a sub-optimal point.
Also, notice that if you start farther away from the optimal point, the gradient descent may take longer to converge.

- Q: Can you identify a case when the gradient descent algorithm would be unable to find the global minimum?
- A: If we choose too large of a learning rate, we might end up over-shooting

## A Non-Convex Example
Now we move on to our non-convex example.

- Q: Can you determine a set of initializations for which the gradient descent algorithm would be stuck in a saddle point?
  - A: $x_0 = [0, \alpha], \; \alpha\in\mathbb{R}$.  

```python
func, grad, grad_flow_step = nonconvex_functions()
title = "NonConvex"

# Parameters
learning_rate = 0.01
num_steps = 20
initial_x = np.random.random(2)  # Initial starting point

# Lists to store the trajectory of x and function values
x_trajectory = [initial_x]
function_values = [func(initial_x)]

# Perform gradient flow steps
for _ in range(num_steps):
    initial_x = grad_flow_step(initial_x, learning_rate)
    x_trajectory.append(initial_x)
    function_values.append(func(initial_x))

# Convert lists to NumPy arrays for easier plotting
x_trajectory = np.array(x_trajectory)
function_values = np.array(function_values)

bounds = np.linspace(-4, 4, 100)
# Plot the trajectory
mesh = np.meshgrid(bounds, bounds)
Z = func(mesh)
plot_steepest_descent_direction(func, grad, title)
plt.contourf(*mesh,
             Z,
             levels=20,
             cmap=plt.cm.jet, alpha=0.6)
plt.scatter(x_trajectory[:, 0], x_trajectory[:, 1], c='red', marker='x', label='Gradient Flow')
plt.plot(x_trajectory[:, 0], x_trajectory[:, 1], c='red', linestyle='dashed', linewidth=0.8)
plt.legend()

# Create a 3D plot
fig = plt.figure(figsize=(7, 3.5))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
surf = ax.plot_surface(*mesh, Z, cmap=plt.cm.jet, rstride=1, cstride=1, alpha=0.8, antialiased=True)
ax.plot(x_trajectory[:, 0], x_trajectory[:, 1], function_values, marker='o', linestyle='-', color='red', label='Gradient Flow Trajectory')

# Add a color bar which maps values to colors
fig.colorbar(surf, shrink=0.5, aspect=5)

# Set labels and title
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('Function Value')
ax.set_title('3D Plot of Non-Convex Function')
```

## An Ill-Conditioned Quadratic Function

- Q: Describe the problem with ill-conditioned problems.
- A:  An ill-conditioned problem is one where, for a small change in the inputs there is a large change in the result. Ill-conditioned problems slow down convergence because high curvature directions require small step-sizes and this leads to slow progress in directions of low curvature. Notice how gradient descent bounces back and forth in high curvature directions. This makes it challenging to find the correct solutions.

- Q: What is the condition number, and what can you say about the condition number of an ill-conditioned quadratic problem?
- A: The condition number of a function measures how much the output value of the function can change for a small change in the input argument. In our quadratic case, the condition number is the ratio of the maximum $\frac{\lambda_\max}{\lambda_\min}$, where $\lambda_i$ are the singular values of $A$.
A problem with a high condition number is said to be ill-conditioned.

- Q: What techniques can we use to proliferate gradient descent in ill-conditioned problems? You can plug and try your solutions in the example snippet below.
  - Pre-conditioning
  - Normalization
  - Modifying the optimization algorithm: momentum or adaptive optimization algorithms (Adam, AdaGrad, etc.). We will cover optimization later in the course, yet you can see intuitively that adapting our step-size for each direction can help mitigate the problem.
  - Gradient Clipping



```python
import numpy as np

eigs = [10, 0.5]
grad_clip, grad_clip_value = False, 2.
# Define an ill-conditioned quadratic function
# x^T A x
def ill_conditioned_quadratic(x):
    return eigs[0] * x[0]**2 + eigs[1] * x[1] ** 2


# Gradient of the ill-conditioned quadratic function
# grad = (A + A^T) x
def gradient_ill_conditioned_quadratic(x):
    return 2 * np.stack((eigs[0] * x[0], eigs[1] * x[1]))

# Generate random initial point
initial_point = np.array([2.0, 2.0])

# Perform gradient descent optimization
learning_rate = 0.1
num_steps = 100
x_trajectory = [initial_point]
function_values = [ill_conditioned_quadratic(initial_point)]

for _ in range(num_steps):
    current_point = x_trajectory[-1]
    gradient = gradient_ill_conditioned_quadratic(current_point)
    if grad_clip:
      gradient = gradient / np.linalg.norm(gradient) * grad_clip_value
    new_point = current_point - learning_rate * gradient
    x_trajectory.append(new_point)
    function_values.append(ill_conditioned_quadratic(new_point))

x_trajectory = np.array(x_trajectory)
function_values = np.array(function_values)

plot_steepest_descent_direction(ill_conditioned_quadratic, gradient_ill_conditioned_quadratic)

color = "red"
plt.scatter(x_trajectory[:, 0], x_trajectory[:, 1], c=color, marker='x', label=f'Gradient Flow {sim}')
plt.plot(x_trajectory[:, 0], x_trajectory[:, 1], c=color, linestyle='dashed', linewidth=0.8)
```
